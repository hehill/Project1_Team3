```python
topics = ['Employment', 'Marital Status','Race','Age','Income','Education','Under 65 Coverage','Sex','Health Care Coverage','Health Care Cost','Depression', 'Healthy Days']
```

```python
new_data = data.loc[data['Topic'].isin(topics)]
new_data = new_data[new_data['Break_Out'] == 'Overall']
new_data.info()

```

```python
new_df = new_data[['Locationabbr'
,'Locationdesc',
'Topic',
'Question',
'Response',
'Sample_Size',
'Data_value',
'GeoLocation']]
new_df.info()
```

```python
reponses = new_df['Response'].nunique()
states = new_df['Locationdesc'].unique()
reponses
```

```python
df_pivot = new_df.pivot_table(index='Locationdesc',columns='Response',values='Data_value')
df_pivot.head()
```

```python
state_df = new_df[['Locationdesc','Locationabbr','GeoLocation']].drop_duplicates()
```

```python
data_cleaning = pd.merge(df_pivot,state_df, on='Locationdesc',how='outer')
data_cleaning.head()
```

```python
data_cleaning.columns
```

```python
data_cleaning.rename(columns={'Locationdesc':'State', 'Yes':'Depression Diagnosis','No':'No Depression Diagnosis','Locationabbr':'State Code'},inplace=True)
data_cleaning.head()
```
